```python
#Import dependencies
import csv
import os
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
from datetime import datetime
from datetime import date
```

```python
#Read in file
file_df = pd.read_csv('GamingStudy_Data.csv', encoding = "ISO-8859-1")

file_df.head()
```

```python
file_df.columns.tolist()
```

```python
#Remove GAD, SWL, SPIN columns and only keep totals

reduced_df = file_df.drop(columns = ['GAD1',
 'GAD2',
 'GAD3',
 'GAD4',
 'GAD5',
 'GAD6',
 'GAD7',
 'SWL1',
 'SWL2',
 'SWL3',
 'SWL4',
 'SWL5','SPIN1',
 'SPIN2',
 'SPIN3',
 'SPIN4',
 'SPIN5',
 'SPIN6',
 'SPIN7',
 'SPIN8',
 'SPIN9',
 'SPIN10',
 'SPIN11',
 'SPIN12',
 'SPIN13',
 'SPIN14',
 'SPIN15',
 'SPIN16',
 'SPIN17','highestleague', "Reference"])

reduced_df.head()
print(len(reduced_df))
```

```python
cleaned_df = reduced_df.dropna(subset=["accept", "GAD_T", "SWL_T","SPIN_T", "Narcissism", 
                                       "GADE", "Hours", "streams", "Work"])

cleaned_df
```

```python
cleaned_df.to_csv("../cleaned_data.csv", index = False, encoding = "ISO-8859-1")
```
